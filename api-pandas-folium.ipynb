## Variables
La variable URL sirve para seleccionar la web donde vamos a hacer el web scraping, la función completa es la siguiente `url = 'https://www.zaragoza.es/sede/servicio/transporte/accidentalidad-trafico/accidente.csv?rows=20'`. La variable coords_zrgz nos permitirá acceder al punto especifico de latitud y longitud que deseamos. Y la variable MAPA nos ilustrará el punto exacto de las coordenadas que escogimos.

```python
url = 'https://www.zaragoza.es/sede/servicio/transporte/accidentalidad-trafico/accidente.csv?rows=100'
coords_zrgz = [41.649693,-0.887712]
mapa = folium.Map(location=coords_zrgz)
```

```python
longitudes = []
latitudes = []
```

```python
longitudes
```

```python
for i in df['geometry']: 
    punto_coord = i.split(',')
    longitudes += [float(punto_coord[1])]
longitudes
```

```python
for i in df['geometry']: 
    punto_coord = i.split(',')
    latitudes += [float(punto_coord[0])]
latitudes
```

```python
df_coord = pd.DataFrame({'long':longitudes,'lat':latitudes})
df_coord
```

```python
df_accidentes = pd.concat([df['type'],df_coord],axis=1)
df_accidentes
```

```python
marcador = folium.Marker(coords_zrgz, popup="¡Hola, Zaragoza!")
mapa = mapa.add_child(marcador)
mapa
```

```python
for indice, fila in df_accidentes.iterrows():
    marcador = folium.Marker([fila['lat'],fila['long']],\
                            popup=fila['type'])
    mapa.add_child(marcador)
mapa
```
